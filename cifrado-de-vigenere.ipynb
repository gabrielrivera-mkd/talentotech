```python
def vigenere_cipher(texto, clave, modo='cifrar'):
    resultado = []
    clave = clave.upper()
    clave_repetida = ''
    i = 0

    for letra in texto:
        if letra.isalpha():
            clave_repetida += clave[i % len(clave)]
            i += 1
        else:
            clave_repetida += letra

    for t, k in zip(texto, clave_repetida):
        if t.isalpha():
            desplazamiento = ord(k) - ord('A')
            if modo == 'descifrar':
                desplazamiento = -desplazamiento

            base = ord('A') if t.isupper() else ord('a')
            letra_cifrada = chr((ord(t) - base + desplazamiento) % 26 + base)
            resultado.append(letra_cifrada)
        else:
            resultado.append(t)

    return ''.join(resultado)

# Ejemplo de uso
texto_original = "payton?"
clave = "ciberseguridad"

texto_cifrado = vigenere_cipher(texto_original, clave, modo='cifrar')
print("Texto cifrado:", texto_cifrado)

texto_descifrado = vigenere_cipher(texto_cifrado, clave, modo='descifrar')
print("Texto descifrado:", texto_descifrado)
```
