## Definir los grafos

### Subtarea
Cree dos gráficos de ejemplo utilizando una biblioteca adecuada como `networkx`.


**Razonamiento**:
Importe la biblioteca networkx y cree dos gráficos con nodos y bordes.



```python
import networkx as nx

# Create the first graph
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (4, 5), (4, 6), (5, 6)])

# Create the second graph
G2 = nx.Graph()
G2.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 1)])
```

## Análisis básico de grafo


### Subtarea:
Para cada grafo, calcular y mostrar propiedades básicas como el número de nodos y aristas, el grado de cada nodo, y verificar si el grafo está conectado.


**Razonamiento**:
Calcule e imprima las propiedades básicas de ambos gráficos como se solicita en las instrucciones.



```python
# Basic properties for G1
print("--- Propiedades de G1 ---")
print(f"Número de nodos: {G1.number_of_nodes()}")
print(f"Número de aristas: {G1.number_of_edges()}")
print("Grado de cada nodo:")
for node, degree in G1.degree():
    print(f"  Nodo {node}: {degree}")
print(f"¿Está conectado?: {nx.is_connected(G1)}")

print("\n")

# Basic properties for G2
print("--- Propiedades de G2 ---")
print(f"Número de nodos: {G2.number_of_nodes()}")
print(f"Número de aristas: {G2.number_of_edges()}")
print("Grado de cada nodo:")
for node, degree in G2.degree():
    print(f"  Nodo {node}: {degree}")
print(f"¿Está conectado?: {nx.is_connected(G2)}")
```

## Análisis de camino/circuito euleriano

### Subtask:
Para cada grafo, determinar si existe un camino o circuito Euleriano y, si es así, encontrar uno.


**Razonamiento**:
Determine si G1 tiene un circuito o camino euleriano y encuentre uno si existe.



```python
print("--- Análisis Euleriano de G1 ---")
is_eulerian_circuit_g1 = nx.is_eulerian(G1)
print(f"¿Tiene circuito Euleriano?: {is_eulerian_circuit_g1}")

if not is_eulerian_circuit_g1:
    has_eulerian_path_g1 = nx.has_eulerian_path(G1)
    print(f"¿Tiene camino Euleriano?: {has_eulerian_path_g1}")
    if has_eulerian_path_g1:
        print("Camino Euleriano encontrado:")
        try:
            eulerian_path_g1 = list(nx.eulerian_path(G1))
            print(eulerian_path_g1)
        except nx.NetworkXError as e:
            print(f"Error al encontrar el camino Euleriano: {e}")
else:
    print("Circuito Euleriano encontrado:")
    try:
        eulerian_circuit_g1 = list(nx.eulerian_circuit(G1))
        print(eulerian_circuit_g1)
    except nx.NetworkXError as e:
        print(f"Error al encontrar el circuito Euleriano: {e}")

print("\n")
```

**Razonamiento**:
Determine si G2 tiene un circuito o camino euleriano y encuentre uno si existe.



```python
print("--- Análisis Euleriano de G2 ---")
is_eulerian_circuit_g2 = nx.is_eulerian(G2)
print(f"¿Tiene circuito Euleriano?: {is_eulerian_circuit_g2}")

if not is_eulerian_circuit_g2:
    has_eulerian_path_g2 = nx.has_eulerian_path(G2)
    print(f"¿Tiene camino Euleriano?: {has_eulerian_path_g2}")
    if has_eulerian_path_g2:
        print("Camino Euleriano encontrado:")
        try:
            eulerian_path_g2 = list(nx.eulerian_path(G2))
            print(eulerian_path_g2)
        except nx.NetworkXError as e:
            print(f"Error al encontrar el camino Euleriano: {e}")
else:
    print("Circuito Euleriano encontrado:")
    try:
        eulerian_circuit_g2 = list(nx.eulerian_circuit(G2))
        print(eulerian_circuit_g2)
    except nx.NetworkXError as e:
        print(f"Error al encontrar el circuito Euleriano: {e}")
```

## Visualización

### Subtask:
Visualizar ambos grafos.


**Razonamiento:**
 importe la biblioteca necesaria, cree figuras y ejes para ambos gráficos, dibuje los gráficos con etiquetas de nodos usando un diseño, establezca títulos y muestre los gráficos.



```python
import matplotlib.pyplot as plt

# Visualize G1
plt.figure(figsize=(8, 6))
pos1 = nx.spring_layout(G1) # or any other layout you prefer
nx.draw(G1, pos1, with_labels=True, node_color='skyblue', node_size=700, edge_color='k', linewidths=1, font_size=15)
plt.title("Visualización del Grafo G1")
plt.axis('off')

# Visualize G2
plt.figure(figsize=(6, 5))
pos2 = nx.spring_layout(G2) # or any other layout you prefer
nx.draw(G2, pos2, with_labels=True, node_color='lightgreen', node_size=700, edge_color='k', linewidths=1, font_size=15)
plt.title("Visualización del Grafo G2")
plt.axis('off')

plt.show()
```
