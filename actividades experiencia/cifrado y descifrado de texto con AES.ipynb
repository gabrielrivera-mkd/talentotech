```python
from cryptography.fernet import Fernet

# Generar y guardar la clave (esto se hace una vez y se comparte de forma segura)
clave = Fernet.generate_key()
cipher = Fernet(clave)

# Mostrar clave (solo para demostración)
print("Clave secreta (compártela con quien descifrará):")
print(clave.decode())

# Mensaje original
mensaje = "Este es un mensaje confidencial."
mensaje_bytes = mensaje.encode()

# Cifrar
mensaje_cifrado = cipher.encrypt(mensaje_bytes)
print("\n Mensaje cifrado:")
print(mensaje_cifrado.decode())

# Simular envío y recepción del mensaje
# Ahora se descifra con la misma clave

cipher_receptor = Fernet(clave)
mensaje_descifrado = cipher_receptor.decrypt(mensaje_cifrado)

print("\n Mensaje descifrado:")
print(mensaje_descifrado.decode())


```
