```python
def es_primo(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            return False
    return True

def primeros_n_primos(n):
    primos = []
    num = 2
    while len(primos) < n:
        if es_primo(num):
            primos.append(num)
        num += 1
    return primos

# Ejemplo de uso
print("Los primeros 25 números primos son:", primeros_n_primos(25))


```
