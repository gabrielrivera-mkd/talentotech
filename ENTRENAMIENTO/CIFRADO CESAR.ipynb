```python
# Cifrado C√©sar b√°sico

def cifrar_cesar(mensaje, desplazamiento):
    resultado = ""
    for letra in mensaje:
        if letra.isalpha():
            base = ord('A') if letra.isupper() else ord('a')
            resultado += chr((ord(letra) - base + desplazamiento) % 26 + base)
        else:
            resultado += letra
    return resultado

def descifrar_cesar(mensaje, desplazamiento):
    return cifrar_cesar(mensaje, -desplazamiento)

# Diccionario con descripci√≥n y precio
catalogo = {
    "delfin": {"desc": "Mam\u00edfero marino inteligente", "precio": 3000},
    "pulpo": {"desc": "Molusco con ocho brazos", "precio": 550},
    "tiburon": {"desc": "Pez depredador", "precio": 5000},
    "ballena": {"desc": "Gigante del oc\u00e9ano", "precio": 10000},
    "estrella de mar": {"desc": "Invertebrado con forma de estrella", "precio": 450},
    "calamar": {"desc": "Molusco de cuerpo blando", "precio": 350},
    "anguila": {"desc": "Pez largo como serpiente", "precio": 200},
    "cangrejo": {"desc": "Cuerpo duro que protege", "precio": 100},
    "orca": {"desc": "Mam\u00edfero marino conocido como ballena asesina", "precio": 2300},
    "pez payaso": {"desc": "Pez peque\u00f1o y colorido", "precio": 250},
    "manta raya": {"desc": "Nada como si volara", "precio": 580},
    "caballito de mar": {"desc": "Pez con forma de caballo", "precio": 2000},
    "medusa": {"desc": "Cuerpo blando y gelatinoso", "precio": 900},
}

carrito = []

while True:
    print("\nüåä BIENVENIDO A LA TIENDA DE ANIMALES MARINOS üåä")
    print("1. Ver cat√°logo")
    print("2. Comprar animal")
    print("3. Ver carrito")
    print("4. Pagar")
    print("5. Cifrar mensaje (Cifrado C√©sar)")
    print("6. Descifrar mensaje (Cifrado C√©sar)")
    print("7. Salir")

    opcion = input("Elige una opci√≥n (1-7): ")

    if opcion == "1":
        for animal, datos in catalogo.items():
            print(f"- {animal.title()}: {datos['desc']} | Precio: ${datos['precio']}")

    elif opcion == "2":
        nombre = input("Ingresa el nombre del animal: ").lower()
        if nombre in catalogo:
            carrito.append(nombre)
            print(f"{nombre.title()} a\u00f1adido al carrito.")
        else:
            print("Ese animal no est\u00e1 en el cat\u00e1logo.")

    elif opcion == "3":
        if carrito:
            total = 0
            print("\nüõí Carrito:")
            for item in carrito:
                precio = catalogo[item]['precio']
                print(f"- {item.title()} : ${precio}")
                total += precio
            print(f"Total: ${total}")
        else:
            print("El carrito est\u00e1 vac\u00edo.")

    elif opcion == "4":
        if carrito:
            print("\nM√©todos de pago disponibles:")
            print("1. Efectivo\n2. Tarjeta\n3. Transferencia")
            metodo = input("Selecciona un m√©todo de pago: ")
            print("¬°Gracias por tu compra! üê†")
            carrito.clear()
        else:
            print("Primero debes agregar animales al carrito.")

    elif opcion == "5":
        mensaje = input("Ingresa el mensaje a cifrar: ")
        despl = int(input("Desplazamiento (clave): "))
        cifrado = cifrar_cesar(mensaje, despl)
        print("Mensaje cifrado:", cifrado)

    elif opcion == "6":
        mensaje = input("Ingresa el mensaje cifrado: ")
        despl = int(input("Desplazamiento (clave): "))
        claro = descifrar_cesar(mensaje, despl)
        print("Mensaje descifrado:", claro)

    elif opcion == "7":
        print("Gracias por visitar nuestra tienda marina üê¨")
        break

    else:
        print("Opci√≥n inv√°lida.")
```
