```python
import itertools
import string

def vigenere_cipher(texto, clave, modo='cifrar'):
    resultado = []
    clave = clave.upper()
    clave_repetida = ''
    i = 0

    for letra in texto:
        if letra.isalpha():
            clave_repetida += clave[i % len(clave)]
            i += 1
        else:
            clave_repetida += letra

    for t, k in zip(texto, clave_repetida):
        if t.isalpha():
            desplazamiento = ord(k) - ord('A')
            if modo == 'descifrar':
                desplazamiento = -desplazamiento

            base = ord('A') if t.isupper() else ord('a')
            letra_cifrada = chr((ord(t) - base + desplazamiento) % 26 + base)
            resultado.append(letra_cifrada)
        else:
            resultado.append(t)

    return ''.join(resultado)

# Texto cifrado a descifrar
texto_cifrado = "clases de payton"  # Este es un ejemplo, reemplázalo por tu texto cifrado

# Fuerza bruta con claves de longitud 2
longitud_clave = 3
abecedario = string.ascii_uppercase
combinaciones = itertools.product(abecedario, repeat=longitud_clave)

print(f"Probando todas las combinaciones de claves de longitud {longitud_clave}:\n")
contador = 0
for clave_tuple in combinaciones:
    clave = ''.join(clave_tuple)
    descifrado = vigenere_cipher(texto_cifrado, clave, modo='descifrar')
    print(f"Clave: {clave} --> Descifrado: {descifrado}")
    contador += 1
    print(f"\nTotal de intentos)  {contador}")

```
