```python
import re
from collections import Counter
import string

# Paso 1: Leer el mensaje cifrado desde un archivo
def leer_mensaje(nombre_archivo):
    with open(nombre_archivo, 'r', encoding='utf-8') as f:
        return f.read()

# Paso 2: Mostrar el mensaje cifrado
def mostrar_mensaje_cifrado(texto):
    print("=== MENSAJE CIFRADO ===")
    print(texto)
    print()

# Paso 3: Análisis de frecuencias (símbolos más comunes)
def analisis_frecuencia(texto):
    texto = texto.replace(" ", "")
    contador = Counter(texto)
    print("=== FRECUENCIA DE SÍMBOLOS ===")
    for simbolo, freq in contador.most_common():
        print(f"'{simbolo}': {freq}")
    print()

# Paso 4: Tabla de símbolos a letras posibles (hipótesis por frecuencia)
def sustituir_simbolos(texto, sustituciones):
    for simbolo, letra in sustituciones.items():
        texto = texto.replace(simbolo, letra)
    return texto

# Paso 5: Vigenère descifrado con clave
def descifrar_vigenere(cifrado, clave):
    abc = string.ascii_uppercase
    cifrado = re.sub(r'[^A-Z]', '', cifrado.upper())
    clave = clave.upper()
    texto_plano = ''
    for i, c in enumerate(cifrado):
        idx_c = abc.index(c)
        idx_k = abc.index(clave[i % len(clave)])
        letra = abc[(idx_c - idx_k) % 26]
        texto_plano += letra
    return texto_plano

# Paso 6: Fuerza bruta con claves de hasta 3 letras (A-Z)
def fuerza_bruta_vigenere(texto, max_largo=3):
    from itertools import product
    abc = string.ascii_uppercase
    texto = re.sub(r'[^A-Z]', '', texto.upper())

    print("=== INTENTOS DE CLAVES ===")
    mejores = []

    with open('/usr/share/dict/spanish', 'r', encoding='utf-8', errors='ignore') as f:
        diccionario = set(palabra.strip().upper() for palabra in f if len(palabra.strip()) > 4)

    for tam in range(1, max_largo+1):
        for clave in product(abc, repeat=tam):
            clave_str = ''.join(clave)
            descifrado = descifrar_vigenere(texto, clave_str)
            palabras = re.findall(r'[A-Z]{5,}', descifrado)
            coincidencias = sum(1 for p in palabras if p in diccionario)
            if coincidencias > 2:
                print(f"Clave: {clave_str} | Palabras válidas: {coincidencias}")
                print(descifrado[:80] + "...")
                mejores.append((clave_str, coincidencias))

    if not mejores:
        print("No se encontraron coincidencias fuertes. Intenta con otra longitud o clave.")
    else:
        print("\n=== POSIBLES CLAVES EXITOSAS ===")
        for clave, match in sorted(mejores, key=lambda x: -x[1])[:5]:
            print(f"Clave: {clave} con {match} coincidencias")

# === USO PRINCIPAL ===

# Cargar mensaje cifrado
archivo = "mensaje.cif"
mensaje = leer_mensaje(archivo)

# Mostrar mensaje original
mostrar_mensaje_cifrado(mensaje)

# Análisis de símbolos
analisis_frecuencia(mensaje)

# Aquí empieza la parte de prueba de claves por fuerza bruta
fuerza_bruta_vigenere(mensaje, max_largo=3)
```
