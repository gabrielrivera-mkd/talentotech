```python
import math

def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def dividir(a, b):
    if b == 0:
        return "‚ùå No se puede dividir por cero"
    return a / b

def log10(a):
    return math.log10(a)

def cos(a):
    return math.cos(a)

def sen(a):
    return math.sin(a)  # CORREGIDO: era math.sen, debe ser math.sin

def tan(a):
    return math.tan(a)

def potencia(a, b):
    return math.pow(a, b)

def raiz(a):
    return math.sqrt(a)

# Funci√≥n principal con ciclo while
def calculadora():
    while True:
        print("\nüî¢ CALCULADORA CIENT√çFICA")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicaci√≥n")
        print("4. Divisi√≥n")
        print("5. Logaritmo (log10)")
        print("6. Coseno")
        print("7. Seno")
        print("8. Tangente")
        print("9. Potencia")
        print("10. Ra√≠z cuadrada")
        print("11. Salir")

        opcion = int(input("Selecciona una opci√≥n (1-11): "))

        if opcion == 11:
            print("üëã ¬°Hasta luego!")
            break  # Salir del ciclo while y terminar el programa

        # Operaciones que usan dos n√∫meros
        if opcion in [1, 2, 3, 4, 9]:
            a = float(input("Ingrese el primer n√∫mero: "))
            b = float(input("Ingrese el segundo n√∫mero: "))
        # Operaciones que usan solo un n√∫mero
        elif opcion in [5, 6, 7, 8, 10]:
            a = float(input("Ingrese el n√∫mero: "))
        else:
            print("‚ùå Opci√≥n inv√°lida.")
            continue

        # Ejecutar operaci√≥n seleccionada
        if opcion == 1:
            print("Resultado:", suma(a, b))
        elif opcion == 2:
            print("Resultado:", resta(a, b))
        elif opcion == 3:
            print("Resultado:", multiplicacion(a, b))
        elif opcion == 4:
            print("Resultado:", dividir(a, b))
        elif opcion == 5:
            print("Resultado:", log10(a))
        elif opcion == 6:
            print("Resultado:", cos(a))
        elif opcion == 7:
            print("Resultado:", sen(a))
        elif opcion == 8:
            print("Resultado:", tan(a))
        elif opcion == 9:
            print("Resultado:", potencia(a, b))
        elif opcion == 10:
            print("Resultado:", raiz(a))

# Ejecutar la calculadora
calculadora()
```

Usamos break cuando el usuario elige salir (opci√≥n 11).
